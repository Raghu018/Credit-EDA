###### AMT_INCOME_TOTAL , AMT_CREDIT are continous variable. So we can create bins for them.

```python
#Creating bins for AMT_INCOME_TOTAL and creating a new column AMT_INCOME_TOTAL_RANGE to show the range of INCOME according to each of their AMT_INCOME_TOTAL.

bins = [0,100000,200000,300000,400000,500000,750000,1000000,10000000000]
slots = ['<100000', '100000-200000','200000-300000','300000-400000','400000-500000','500000-750000','750000-1000000', '1000000 and above']

inp0['AMT_INCOME_TOTAL_RANGE'] = pd.cut(inp0['AMT_INCOME_TOTAL'], bins = bins, labels=slots)
```
